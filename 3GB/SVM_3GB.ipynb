```python
# this code load 20000 image only
from PIL import Image
import os, sys
import cv2
import numpy as np

# load all images in a directory
from os import listdir
from matplotlib import image
# load all images in a directory
loaded_images = []
images_label = []
i = 0
path='/content/IDC_regular_ps50_idx5'
for big_folder in listdir(path):
    for folder in listdir(path + '/' + big_folder  ):
        for filename in listdir(path + '/' + big_folder + '/' + folder):
          if i > 20000:
            break
          # load image
          img_data = Image.open(path + '/' + big_folder +'/' + folder + '/' +filename).convert('RGB')
          if np.array(img_data).shape !=(50, 50, 3):
             continue
          else:
             img_data = np.array(img_data)
             #img_data = image.imread(path + '/' + big_folder +'/' + folder + '/' +filename)
             # store loaded image
             loaded_images.append(img_data)
             images_label.append(folder)
             i +=1
             #print('> loaded %s %s' % (filename, img_data.shape))
```

```python
arr_images=np.array(loaded_images)
```

```python
test_ratio = 0.25
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(arr_images, arr_labels, test_size=test_ratio, random_state=1, shuffle=True)
print("the X_train shape", X_train.shape)
print("the X_test shape", X_test.shape) 
```

```python
nsamples, nx, ny, dim = X_train.shape
d2_X_train = X_train.reshape((nsamples, nx*ny*dim))
nsamples1, nx1, ny1, dim1 = X_test.shape
d2_X_test = X_test.reshape((nsamples1, nx1*ny1*dim1))
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(d2_X_train)
X_test = sc.transform(d2_X_test)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', random_state=0)
print(classifier.fit(X_train, y_train))
```

```python
y_pred = classifier.predict(X_test)
```
